```python
import pandas as pd
```

```python
df = pd.read_csv('../large_order_data_2024_10M.csv')
df
```

```python
print(df.shape)
print(df.dtypes)
print(df['transaction_datetime'].min(), df['transaction_datetime'].max())
```

```python
# Convert transaction_datetime to datetime
df['transaction_datetime'] = pd.to_datetime(df['transaction_datetime'], errors='coerce')

# เพิ่มคอลัมน์เดือน
df['month'] = df['transaction_datetime'].dt.to_period('M')
df.head()
```

```python
df
```

```python
# คำนวณยอดใช้จ่ายรวมของแต่ละ customer ต่อเดือน
monthly_spend = (
    df.groupby(['month', 'customer_no'], as_index=False)['amount']
      .sum()
      .rename(columns={'amount': 'total_spent'})
)
monthly_spend
```

```python
# หาลูกค้า Top 10 spender ของแต่ละเดือน
top_spenders = (
    monthly_spend
    .sort_values(['month', 'total_spent'], ascending=[True, False])
    .groupby('month')
    .head(10)
)

# เพิ่ม rank
top_spenders['rank'] = top_spenders.groupby('month')['total_spent'].rank(method='first', ascending=False).astype(int)

top_spenders
```

```python
top_spenders.to_csv('../top spender.csv', index=False)
```
